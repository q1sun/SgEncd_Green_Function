```python
##############################################################################################
import torch
import torch.nn as nn
import numpy as np
import os
import time
import datetime
import argparse 

from torch import optim, autograd
from matplotlib import pyplot as plt

# matlab to tensor and vice versa
from scipy.io import loadmat
import numpy as np
import scipy.io as io

# create training and testing datasets
from torch.utils.data import Dataset, DataLoader
from DataSets.Line1D import Sample_Points, Exact_Solution
from Utils import helper

# create neural network surrogate model
from Models.FcNet import FcNet

# load data from two datasets within the same loop
from itertools import cycle

print("pytorch version", torch.__version__, "\n")

## parser arguments
parser = argparse.ArgumentParser(description='Deep Green Function for 1D Poisson Equation using PINNs')
# checkpoints
parser.add_argument('-c', '--checkpoint', default='Checkpoints/Poisson1D/simulation_0', type=str, metavar='PATH', help='path to save checkpoint')
# datasets
parser.add_argument('-i', '--image', default='Images/Poisson1D/simulation_0', type=str, metavar='PATH', help='path to save figures')
args, unknown = parser.parse_known_args()
# args = parser.parse_args()
##############################################################################################



##############################################################################################
## hyperparameter configuration
## ------------------------- ##
# problem setting
dim_prob = 1
dim_in = 2
## ------------------------- ##
# dataset setting
num_intrr_pts_x = 300
num_intrr_pts_y = 300
num_bndry_pts_y = 2 # default for 1D problem
num_test_pts = 100 

batch_num = 10
batchsize = num_intrr_pts_y * num_intrr_pts_y // batch_num
## ------------------------- ##
# network setting
width = 40
depth = 3
## ------------------------- ##
# optimization setting
beta_bndry = 800 # penalty coefficient for Dirichlet boundary
beta_symtr = 800 # penalty coefficient for symmetry of Green's function
num_epochs = 650
milestones = 300, 500, 600
##############################################################################################



##############################################################################################
print('*', '-' * 45, '*')
print('===> preparing training and testing datasets ...')

# training dataset for sample points inside the domain
class Traindata(Dataset):    
    def __init__(self, num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob): 
        
        # SmpPts_Train[ x-SmpPt, y-SmpPt, 0, 1 ], each x-SmpPt aligns with all y-SmpPts inside (0,1)
        self.SmpPts_Train = Sample_Points.SmpPts_Train_Line1D(num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob)
        
        self.f_Intrr_SmpPts = Exact_Solution.f_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,1])[:, None]      
        self.g_Bndry_SmpPts = torch.cat( (Exact_Solution.g_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,2])[:, None], 
                                          Exact_Solution.g_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,3])[:, None]), 1)                        
            
    def __len__(self):
        return len(self.SmpPts_Train)
    
    def __getitem__(self, idx):
        SmpPt = self.SmpPts_Train[idx]
        f_SmpPt = self.f_Intrr_SmpPts[idx]
        g_SmpPt = self.g_Bndry_SmpPts[idx]

        return [SmpPt, f_SmpPt, g_SmpPt]
    
# testing dataset for equidistant sample points over the entire domain
class Testdata(Dataset):    
    def __init__(self, num_test_pts): 
        
        self.SmpPts_Test = Sample_Points.SmpPts_Test_Square2D(num_test_pts)
        self.G_Exact_SmpPts = Exact_Solution.Green_Exact_Line1D(self.SmpPts_Test[:,0], self.SmpPts_Test[:,1])
        self.F_Exact_SmpPts = Exact_Solution.FdmSlt_Exact_Line1D(self.SmpPts_Test[:,0], self.SmpPts_Test[:,1])        
              
    def __len__(self):
        return len(self.SmpPts_Test)
    
    def __getitem__(self, idx):
        SmpPt = self.SmpPts_Test[idx]
        G_Exact_SmpPt = self.G_Exact_SmpPts[idx]
        F_Exact_SmpPt = self.F_Exact_SmpPts[idx]
 
        return [SmpPt, G_Exact_SmpPt, F_Exact_SmpPt] 

# create training and testing datasets         
traindata = Traindata(num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob)
testdata = Testdata(num_test_pts)

# define dataloader 
dataloader_train = DataLoader(traindata, batch_size=batchsize, shuffle=True, num_workers=0)
dataloader_test = DataLoader(testdata, batch_size=num_test_pts*num_test_pts, shuffle=False, num_workers=0)

print('===> done!')
print('*', '-' * 45, '*')
##############################################################################################
```
