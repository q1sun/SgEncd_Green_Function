```python
##############################################################################################
import torch
import torch.nn as nn
import numpy as np
import os
import time
import datetime
import argparse 

from torch import optim, autograd
from matplotlib import pyplot as plt

from scipy.io import loadmat
import numpy as np
import scipy.io as io

# create training and testing datasets
from torch.utils.data import Dataset, DataLoader
from DataSets.Line1D import Sample_Points, Exact_Solution
from Utils import helper

# create neural network surrogate model
from Models.FcNet import FcNet

# load data from two datasets within the same loop
from itertools import cycle

print("pytorch version", torch.__version__, "\n")

## parser arguments
parser = argparse.ArgumentParser(description='Deep Green Function for 1D Poisson Equation using PINNs')
# checkpoints
parser.add_argument('-c', '--checkpoint', default='Checkpoints/Poisson1D/simulation_0', type=str, metavar='PATH', help='path to save checkpoint')
# datasets
parser.add_argument('-i', '--image', default='Images/Poisson1D/simulation_0', type=str, metavar='PATH', help='path to save figures')
args, unknown = parser.parse_known_args()
# args = parser.parse_args()
##############################################################################################



##############################################################################################
## hyperparameter configuration
## ------------------------- ##
# problem setting
dim_prob = 1
dim_in = 2
## ------------------------- ##
# dataset setting
num_intrr_pts_x = 100
num_intrr_pts_y = 100
num_bndry_pts_y = 2 # default for 1D problem
num_test_pts = 100 

batch_num = 16
batchsize = num_intrr_pts_y * num_intrr_pts_y // batch_num
## ------------------------- ##
# network setting
width = 40
depth = 3
## ------------------------- ##
# optimization setting
beta = 400 # intial penalty coefficient
num_epochs = 5
milestones = 2, 3
##############################################################################################



##############################################################################################
print('*', '-' * 45, '*')
print('===> preparing training and testing datasets ...')

# training dataset for sample points inside the domain
class Traindata(Dataset):    
    def __init__(self, num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob): 
        
        # SmpPts_Train[ x-SmpPt, y-SmpPt, 0, 1 ], each x-SmpPt aligns with all y-SmpPts inside (0,1)
        self.SmpPts_Train = Sample_Points.SmpPts_Train_Line1D(num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob)
        
        self.f_Intrr_SmpPts = Exact_Solution.f_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,1])[:, None]      
        self.g_Bndry_SmpPts = torch.cat( (Exact_Solution.g_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,2])[:, None], 
                                          Exact_Solution.g_Exact_Line1D(self.SmpPts_Train[:,0], self.SmpPts_Train[:,3])[:, None]), 1)                        
            
    def __len__(self):
        return len(self.SmpPts_Train)
    
    def __getitem__(self, idx):
        SmpPt = self.SmpPts_Train[idx]
        f_SmpPt = self.f_Intrr_SmpPts[idx]
        g_SmpPt = self.g_Bndry_SmpPts[idx]

        return [SmpPt, f_SmpPt, g_SmpPt]
    
# testing dataset for equidistant sample points over the entire domain
class Testdata(Dataset):    
    def __init__(self, num_test_pts): 
        
        self.SmpPts_Test = Sample_Points.SmpPts_Test_Square2D(num_test_pts)
        self.G_Exact_SmpPts = Exact_Solution.Green_Exact_Line1D(self.SmpPts_Test[:,0], self.SmpPts_Test[:,1])
        self.F_Exact_SmpPts = Exact_Solution.FdmSlt_Exact_Line1D(self.SmpPts_Test[:,0], self.SmpPts_Test[:,1])        
              
    def __len__(self):
        return len(self.SmpPts_Test)
    
    def __getitem__(self, idx):
        SmpPt = self.SmpPts_Test[idx]
        G_Exact_SmpPt = self.G_Exact_SmpPts[idx]
        F_Exact_SmpPt = self.F_Exact_SmpPts[idx]
 
        return [SmpPt, G_Exact_SmpPt, F_Exact_SmpPt] 

# create training and testing datasets         
traindata = Traindata(num_intrr_pts_x, num_intrr_pts_y, num_bndry_pts_y, dim_prob)
testdata = Testdata(num_test_pts)

# define dataloader 
dataloader_train = DataLoader(traindata, batch_size=batchsize, shuffle=True, num_workers=0)
dataloader_test = DataLoader(testdata, batch_size=num_test_pts*num_test_pts, shuffle=False, num_workers=0)

print('===> done!')
print('*', '-' * 45, '*')
##############################################################################################



##############################################################################################
# plot sample points during training and testing
if not os.path.isdir(args.image):
    helper.mkdir_p(args.image)
    
fig = plt.figure()
plt.scatter(traindata.SmpPts_Train[:,0], traindata.SmpPts_Train[:,1], c = 'red', label = 'interior points' )
plt.scatter(traindata.SmpPts_Train[:,0], traindata.SmpPts_Train[:,2], c = 'blue', label = 'boundary points' )
plt.scatter(traindata.SmpPts_Train[:,0], traindata.SmpPts_Train[:,3], c = 'blue' )
plt.title('Sample Points during Training')
plt.legend(loc = 'lower right')
# plt.show()
fig.savefig(os.path.join(args.image,'TrainSmpPts.png'))

fig = plt.figure()
plt.scatter(testdata.SmpPts_Test[:,0], testdata.SmpPts_Test[:,1], c = 'black')
plt.title('Sample Points during Testing')
# plt.show()
fig.savefig(os.path.join(args.image,'TestSmpPts.png'))
##############################################################################################



##############################################################################################
print('*', '-' * 45, '*')
print('===> creating training model ...')
print('*', '-' * 45, '*', "\n", "\n")

def train_epoch(epoch, model, optimizer, device):
    
    # set model to training mode
    model.train()

    loss_epoch, loss_intrr_epoch, loss_bndry_epoch = 0, 0, 0

    for i, data_minibatch in enumerate(dataloader_train):
        
        # get mini-batch training data
        smppts, f_smppts, g_smppts = data_minibatch
        smppts_intrr = smppts[:,[0,1]]
        smppts_bndryL = smppts[:,[0,2]] # left
        smppts_bndryR = smppts[:,[0,3]] # right 

        # send training data to device            
        smppts_intrr = smppts_intrr.to(device)
        smppts_bndryL = smppts_bndryL.to(device)
        smppts_bndryR = smppts_bndryR.to(device) 
        f_smppts = f_smppts.to(device)
        g_smppts = g_smppts.to(device)
        
        smppts_intrr.requires_grad = True
        
        # forward pass to obtain NN prediction of correction function H(x,y)
        NN_intrr = model(smppts_intrr)
        NN_bndryL = model(smppts_bndryL)
        NN_bndryR = model(smppts_bndryR)
        # zero parameter gradients and then compute NN prediction of {d^2/dy^2} H(x,y)
        model.zero_grad()
        grad_NN_intrr = torch.autograd.grad(outputs=NN_intrr, inputs=smppts_intrr, grad_outputs = torch.ones_like(NN_intrr), retain_graph=True, create_graph=True, only_inputs=True)[0]                                        
        grad_NN_dy_intrr = torch.autograd.grad(outputs=torch.squeeze(grad_NN_intrr[:,1]), inputs=smppts_intrr, grad_outputs = torch.ones_like(grad_NN_intrr[:,1]), retain_graph=True, create_graph=True, only_inputs=True)[0]        
                
        # construct mini-batch loss function and then perform backward pass
        loss_intrr = torch.mean( torch.pow( grad_NN_dy_intrr[:,1] + f_smppts, 2 ) )
        loss_bndryL = torch.mean( torch.pow( torch.squeeze(NN_bndryL) - g_smppts[:,0], 2))
        loss_bndryR = torch.mean( torch.pow( torch.squeeze(NN_bndryR) - g_smppts[:,1], 2))
                                       
        loss_minibatch = loss_intrr + beta * (loss_bndryL + loss_bndryR)

        # zero parameter gradients
        optimizer.zero_grad()
        # backpropagation
        loss_minibatch.backward()
        # parameter update
        optimizer.step()     

        # integrate loss over the entire training datset
        loss_intrr_epoch += loss_intrr.item() * smppts.size(0) / traindata.SmpPts_Train.shape[0]
        loss_bndry_epoch += ( loss_bndryL.item() + loss_bndryR.item() ) * smppts.size(0) / traindata.SmpPts_Train.shape[0]
        loss_epoch += loss_intrr_epoch + beta * loss_bndry_epoch

    return loss_epoch, loss_intrr_epoch, loss_bndry_epoch
##############################################################################################



##############################################################################################
print('*', '-' * 45, '*')
print('===> creating testing model ...')
print('*', '-' * 45, '*', "\n", "\n")

def test_epoch(epoch, model, optimizer, device):
    
    # set model to testing mode
    model.eval()

    loss_epoch = 0

    for smppts_test, G_exact_smppts, F_exact_smppts in dataloader_test:
        
        # send inputs, outputs to device
        smppts_test = smppts_test.to(device)
        G_exact_smppts = G_exact_smppts.to(device)
        F_exact_smppts = F_exact_smppts.to(device)
        
        # forward pass and then compute loss function for approximating G(x,y) by H_NN(x,y) + F(x,y)
        NN_smppts = model(smppts_test) 
        
        loss_epoch = torch.mean( torch.pow( torch.squeeze(NN_smppts) + F_exact_smppts - G_exact_smppts, 2 ) )                 
                
        # integrate loss      
        loss_epoch += loss_epoch.item()          
    
    return loss_epoch
##############################################################################################


##############################################################################################
print('*', '-' * 45, '*')
print('===> training neural network ...')

if not os.path.isdir(args.checkpoint):
    helper.mkdir_p(args.checkpoint)

# create model
model = FcNet.FcNet(dim_in, width, 1, depth)
model.Xavier_initi()
print('Network Architecture:', "\n", model)
print('Total number of trainable parameters = ', sum(p.numel() for p in model.parameters() if p.requires_grad))

# create optimizer and learning rate schedular
optimizer = torch.optim.AdamW(model.parameters(), lr=0.01, betas=(0.9, 0.999), eps=1e-08, weight_decay=0.01, amsgrad=False)
schedular = torch.optim.lr_scheduler.MultiStepLR(optimizer, milestones, gamma=0.1)

# load model to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print('DEVICE: {}'.format(device), "\n")
model = model.to(device)

# create log file
logger = helper.Logger(os.path.join(args.checkpoint, 'log.txt'), title='DeepGreen-Poisson-Line1D')
logger.set_names(['Learning Rate', 'TrainLoss', 'TestLoss', 'TrainL(intrr)', 'TrainL(bndry)'])
     
# train and test 
trainloss_curve, testloss_curve = [], []
trainloss_best = 1e10
since = time.time()
for epoch in range(num_epochs):
      
    print('Epoch {}/{}'.format(epoch, num_epochs-1), 'with LR = {:.1e}'.format(optimizer.param_groups[0]['lr']))  
        
    # execute training and testing
    trainloss_epoch, trainloss_intrr_epoch, trainloss_bndry_epoch  = train_epoch(epoch, model, optimizer, device)
    testloss_epoch = test_epoch(epoch, model, optimizer, device)
    
    # save current and best models to checkpoint
    is_best = trainloss_epoch < trainloss_best
    if is_best:
        print('==> Saving best model ...')
    trainloss_best = min(trainloss_epoch, trainloss_best)
    helper.save_checkpoint({'epoch': epoch + 1,
                            'state_dict': model.state_dict(),
                            'trainloss_intrr_epoch': trainloss_intrr_epoch,
                            'trainloss_bndry_epoch': trainloss_bndry_epoch,
                            'trainloss_epoch': trainloss_epoch,
                            'testloss_epoch': testloss_epoch,
                            'trainloss_best': trainloss_best,
                            'optimizer': optimizer.state_dict(),
                           }, is_best, checkpoint=args.checkpoint)   
    # save training process to log file
    logger.append([optimizer.param_groups[0]['lr'], trainloss_epoch, testloss_epoch, trainloss_intrr_epoch, trainloss_bndry_epoch])
    
    # adjust learning rate according to predefined schedule
    schedular.step()

    # print results
    trainloss_curve.append(trainloss_epoch)
    testloss_curve.append(testloss_epoch)
    print('==> Full-Batch Training Loss = {:.4e}'.format(trainloss_epoch), '  Testing Loss = {:.4e}'.format(testloss_epoch), "\n")
    
logger.close()
time_elapsed = time.time() - since

print('Done in {}'.format(str(datetime.timedelta(seconds=time_elapsed))), '!')
print('*', '-' * 45, '*', "\n", "\n")
##############################################################################################



##############################################################################################
# plot learning curves
fig = plt.figure()
plt.plot(trainloss_curve, c = 'red', label = 'training loss' )
plt.title('Learning Curve during Training')
plt.legend(loc = 'upper right')
# plt.show()
fig.savefig(os.path.join(args.image,'TrainCurve.png'))

fig = plt.figure()
plt.plot(torch.log10(torch.tensor(testloss_curve)), c = 'red', label = 'testing loss' )
plt.title('Learning Curve during Testing')
plt.legend(loc = 'upper right')
# plt.show()
fig.savefig(os.path.join(args.image,'TestCurve.png'))
##############################################################################################



##############################################################################################
print('*', '-' * 45, '*')
print('===> loading trained model for inference ...')

# load trained model
checkpoint = torch.load(os.path.join(args.checkpoint, 'model_best.pth.tar'))
model.load_state_dict(checkpoint['state_dict'])

# compute NN predicution of correction function H(x,y)
with torch.no_grad():    
    SmpPts_Test = testdata.SmpPts_Test
    SmpPts_Test = SmpPts_Test.to(device)
    NN_Test = model(SmpPts_Test) 

# plot results
x = testdata.SmpPts_Test[:,0].reshape(num_test_pts,num_test_pts)
y = testdata.SmpPts_Test[:,1].reshape(num_test_pts,num_test_pts)

with torch.no_grad(): 

    # plot exact solution using testing sample points
    plt.figure()
    G_Exact = testdata.G_Exact_SmpPts.reshape(num_test_pts,num_test_pts)
    plt.contourf(x, y, G_Exact, 100, cmap = 'jet')
    plt.title('Exact Green Function on Test Dataset')
    bounds = torch.linspace(G_Exact.min(), G_Exact.max(), 5)
    plt.colorbar(ticks=bounds)#(orientation='horizontal')
    # plt.show()  
    plt.savefig(os.path.join(args.image,'Green_Exact_TestData.png'))
    plt.close(fig)


    plt.figure()
    F_Exact = testdata.F_Exact_SmpPts.reshape(num_test_pts,num_test_pts)
    H_NN = NN_Test.reshape(num_test_pts, num_test_pts)  
    G_NN = H_NN + F_Exact
    plt.contourf(x, y, G_NN, 40, cmap = 'jet')
    plt.title('Deep Green Function on Test Dataset')
    bounds = torch.linspace(G_NN.min(), G_NN.max(), 5)
    plt.colorbar(ticks=bounds)#(orientation='horizontal')
    # plt.show()  
    plt.savefig(os.path.join(args.image,'Green_Deep_TestData.png'))
    plt.close(fig)

    plt.figure()
    PtErr = G_Exact - G_NN
    plt.contourf(x, y, PtErr, 40, cmap = 'jet')
    plt.title('Pointwise Error on Test Dataset')
    plt.colorbar()
    # plt.show() 
    plt.savefig(os.path.join(args.image,'Green_PtErr_TestData.png'))
    plt.close(fig)        
        
print('*', '-' * 45, '*', "\n", "\n")
##############################################################################################





















```
